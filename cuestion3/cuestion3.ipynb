```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D


datos=pd.read_csv("https://raw.githubusercontent.com/ernesto106/HitoIndividual1-SGE_Ernesto/main/cuestion3/IMDB-Movie-Data.csv")
df=pd.DataFrame(datos)

# Gráfica de pastel que muestra el total de películas por años. Como vemos, el año 2016 es el que mas películas contiene.

df.Rank.groupby(df.Year).sum().plot(kind="pie", cmap="Paired", autopct='%0.1f%%')
plt.ylabel('')
plt.title('Total de películas por años')
```

```python
"""

Esta opción es muy interesante, pues se crean  varias gráficas que  muestran por años la correlación existente 
entre el número de votos y la calificación que ha recibido la película. 
Como vemos, cuanto mayor es el número de votos (Votes) que ha recibido, mayor es su calificación (Rating).


"""

g=sns.lmplot(x='Votes', y='Rating', data=df, col='Year', aspect=.4,x_jitter=0.1)
```

```python
"""

Gráfica que muestra la correlación existente entre los votos que ha recibido la película y su presupuesto.
Como vemos, hay muchas películas con un presupuesto bajo y menor cantidad de votos. Sin embargo, 
hay pocas películas con un presupuesto alto, pero cuanto mayor es el presupuesto invertido en la pelicula, 
se aprecia que ha recibido una cantidad de votos mayor. 

"""

g=sns.lmplot(x='Revenue (Millions)', y='Votes', data=df,palette='Set1')

plt.ylabel("votos")
plt.xlabel("presupuesto")
plt.title('Correlación existente entre el presupuesto de las películas y su número de votos')
```

```python
"""

Gráfica en 3D que muestra la misma correlación anterior, pero además podemos identificar gracias al eje 'y' el id de la película, 
con el total de las más de 1000 películas que contiene el archivo. 
Cuanto mayor el presupuesto, menor cantidad de peliculas, pero más votos reciben.
Cuanto menor el presupuesto, mayor cantidad de peliculas, pero menos votos reciben.

"""

x=datos['Votes']
y=datos['Revenue (Millions)']
z=datos['Rank']

fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.scatter3D(x, y, z, c=z, cmap='Set1')
ax.set_xlabel("votos")
ax.set_ylabel("presupuesto")
ax.set_zlabel("id_pelicula")
plt.show()
```
